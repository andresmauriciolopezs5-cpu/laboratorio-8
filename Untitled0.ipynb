```python
import random

def generar_numero():
    """Genera un número aleatorio entre 1 y 100."""
    return random.randint(1, 100)

def juego_adivinar_numero():
    print("¡Bienvenido al juego de adivinar el número!")
    print("Estoy pensando en un número entre 1 y 100.")

    numero_secreto = generar_numero()
    intentos = 0

    while True:
        try:
            entrada_usuario = input("Adivina el número (entre 1 y 100): ")
            intento_jugador = int(entrada_usuario)
            intentos += 1

            if intento_jugador < 1 or intento_jugador > 100:
                print("Por favor, introduce un número entre 1 y 100.")
                continue

            if intento_jugador < numero_secreto:
                print("¡Incorrecto! El número secreto es mayor.")
            elif intento_jugador > numero_secreto:
                print("¡Incorrecto! El número secreto es menor.")
            else:
                print(f"¡Felicidades! ¡Adivinaste el número {numero_secreto} en {intentos} intentos!")
                break
        except ValueError:
            print("Entrada inválida. Por favor, introduce un número entero.")

# Para iniciar el juego, simplemente llama a la función:
juego_adivinar_numero()
```

```python
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        raise ZeroDivisionError("No se puede dividir por cero.")
    return a / b

def calculadora():
    while True:
        print("\n--- Calculadora Simple ---")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")
        print("5. Salir")

        opcion = input("Selecciona una operación (1-5): ")

        if opcion == '5':
            print("¡Hasta luego!")
            break

        if opcion in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Introduce el primer número: "))
                num2 = float(input("Introduce el segundo número: "))
            except ValueError:
                print("Entrada inválida. Por favor, introduce números válidos.")
                continue

            resultado = None
            if opcion == '1':
                resultado = suma(num1, num2)
            elif opcion == '2':
                resultado = resta(num1, num2)
            elif opcion == '3':
                resultado = multiplicacion(num1, num2)
            elif opcion == '4':
                try:
                    resultado = division(num1, num2)
                except ZeroDivisionError as e:
                    print(e)
                    continue

            if resultado is not None:
                print(f"El resultado es: {resultado}")
        else:
            print("Opción inválida. Por favor, selecciona una opción entre 1 y 5.")

# Para iniciar la calculadora, simplemente llama a la función:
calculadora()
```

```python
lista_de_tareas = []

def agregar_tarea(tarea):
    lista_de_tareas.append(tarea)
    print(f"Tarea '{tarea}' agregada.")

def eliminar_tarea(indice):
    try:
        tarea_eliminada = lista_de_tareas.pop(indice - 1)
        print(f"Tarea '{tarea_eliminada}' eliminada.")
    except IndexError:
        print("Error: Índice de tarea inválido. Por favor, introduce un número de tarea existente.")
    except ValueError:
        print("Entrada inválida. Por favor, introduce un número.")

def mostrar_tareas():
    if not lista_de_tareas:
        print("No hay tareas pendientes.")
    else:
        print("\n--- Tus Tareas ---")
        for i, tarea in enumerate(lista_de_tareas):
            print(f"{i + 1}. {tarea}")
        print("-------------------")

def gestor_tareas():
    while True:
        print("\n--- Gestor de Tareas ---")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Mostrar tareas")
        print("4. Salir")

        opcion = input("Selecciona una opción (1-4): ")

        if opcion == '1':
            tarea = input("Introduce la nueva tarea: ")
            agregar_tarea(tarea)
        elif opcion == '2':
            mostrar_tareas()
            if lista_de_tareas:
                try:
                    indice_a_eliminar = int(input("Introduce el número de la tarea a eliminar: "))
                    eliminar_tarea(indice_a_eliminar)
                except ValueError:
                    print("Entrada inválida. Por favor, introduce un número entero.")
        elif opcion == '3':
            mostrar_tareas()
        elif opcion == '4':
            print("¡Hasta luego!")
            break
        else:
            print("Opción inválida. Por favor, selecciona una opción entre 1 y 4.")

# Para iniciar el gestor de tareas, simplemente llama a la función:
gestor_tareas()
```
